```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn3

# Definir subconjuntos com tamanhos
venn = venn3(
    subsets={'100': 10, '010': 10, '001': 10,
             '110': 10, '101': 10, '011': 10, '111': 15},
    set_labels=('RF', 'XGBoost', 'MLP')
)

# Adicionar listas de strings às interseções
# Somente no conjunto A
venn.get_label_by_id('100').set_text("N012\nN015\nD00901")

# Somente no conjunto B
venn.get_label_by_id('010').set_text("VDF002\nN00101")

# Somente no conjunto C
venn.get_label_by_id('001').set_text("C009_2\nC011_1.0\nC011_4.0")

# Interseção entre A e B
venn.get_label_by_id('110').set_text("C00703")

# Interseção entre A e C
venn.get_label_by_id('101').set_text("")

# Interseção entre B e C
venn.get_label_by_id('011').set_text("H001")

# Interseção entre A, B e C
venn.get_label_by_id('111').set_text("J007, N016\nC006, N010\nN017, N013")

# Configurar título e exibição
plt.title("Atributos Mais Relevantes")
plt.show()
```
