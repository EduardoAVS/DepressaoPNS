```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pickle
with open('dados/depresso.pkl', 'rb') as f:
  X_prev, X_test, y_classe, y_test = pickle.load(f)
```

```python
param_distributions = {
    'C': [0.502],  # Valores para o parâmetro de regularização
    'kernel': ['rbf'],  # Diferentes tipos de kernels
    'gamma': ['auto'],  # Para kernels rbf, poly e sigmoid
    'degree': [2]  # Para o kernel polinomial
}
```

```python
svm_model = SVC()
```

```python
modelo = RandomizedSearchCV(estimator=svm_model, param_distributions=param_distributions, 
                                   n_iter=20, cv=3, verbose=1, random_state=42, n_jobs=-1)
```

```python
modelo.fit(X_prev, y_classe)
```

```python

print(modelo.best_params_)
```

```python
melhor_modelo = modelo.best_estimator_
previsoes = melhor_modelo.predict(X_test)
```

```python
from yellowbrick.classifier import ConfusionMatrix
confusion_matrix(y_test, previsoes)
```

```python
cm = ConfusionMatrix(modelo)
cm.fit(X_prev, y_classe)
cm.score(X_test, y_test)
```

```python
print(classification_report(y_test, previsoes))
```
